# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
#Separate the data into 3 groups: small, med, large
small_df = clean_df[(clean_df['Number of Discharges']<100)]
small_df = small_df.sort_values('Excess Readmission Ratio')
print('small hospitals # of discharges: ',sum(small_df['Number of Discharges']))
print('small hospitals sum of excess readmission ratio: ', sum(small_df['Excess Readmission Ratio']))
med_df = clean_df[((clean_df['Number of Discharges']>=100) 
                               | (clean_df['Number of Discharges']<1000))]
med_df = med_df.sort_values('Excess Readmission Ratio')
large_df = clean_df[(clean_df['Number of Discharges']>=1000)]
large_df = large_df.sort_values('Excess Readmission Ratio')
```

```python
# EDA: show ECDF for small vs large hospitals
# Descriptive statistics 
print('Large Hospitals')
print(large_df['Excess Readmission Ratio'].describe())
print('')
print('Small Hospitals')
print(small_df['Excess Readmission Ratio'].describe())

large_mean = np.mean(large_df['Excess Readmission Ratio'])
small_mean = np.mean(small_df['Excess Readmission Ratio'])
print('large mean: ', large_mean,'small mean: ',small_mean,'difference: ',large_mean-small_mean)

# Compute ecdfs
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)
    # x-data for the ECDF: x
    x = np.sort(data)
    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n
    return x, y

x_large, y_large = ecdf(large_df['Excess Readmission Ratio'])
x_small, y_small = ecdf(small_df['Excess Readmission Ratio'])

# Examine differences visually
fig = plt.figure(figsize=(12, 12))
_ = plt.subplot(2,1,1)
_ = plt.hist(large_df['Excess Readmission Ratio'],bins=30, normed=True, histtype='step')
_ = plt.hist(small_df['Excess Readmission Ratio'],bins=30, normed=True, histtype='step')
_ = plt.xlabel('Excesss Readmission Ratio')
_ = plt.ylabel('PDF')
_ = plt.title('Readmission Ratio According to Hosptial Size')
_ = plt.legend(('Large hospitals', 'Small hospitals'))

_ = plt.subplot(2,1,2)
_ = plt.plot(x_large, y_large, marker='.', linestyle='none')
_ = plt.plot(x_small, y_small, marker='.', linestyle='none')
_ = plt.ylabel('ECDF')
_ = plt.xlabel('Excesss Readmission Ratio')
_ = plt.title('Readmission Ratio According to Hosptial Size')
_ = plt.legend(('Large hospitals', 'Small hospitals'))
```

```python
# Null hypothesis: The distributions for large and small hospitals are identical.
# Define test statistic 
def diff_of_means(data_1, data_2):
    """Difference in means of two arrays"""
    return np.mean(data_1) - np.mean(data_2)

# Compute test statistic for observed data
obs_diff = diff_of_means(large_df['Excess Readmission Ratio'], small_df['Excess Readmission Ratio'])

# Draw permutation replicates
perm_replicates = np.empty(10000)
for i in range(len(perm_replicates)): 
    # Permute the data
    both = np.concatenate((large_df['Excess Readmission Ratio'], small_df['Excess Readmission Ratio']))
    both_perm = np.random.permutation(both)
    perm_large = both_perm[:len(large_df)]
    perm_small = both_perm[len(large_df):]
    #Draw replicates
    perm_replicates[i] = diff_of_means(perm_large, perm_small)

# Compute the p-value
p_value = np.sum(abs(perm_replicates) >= abs(obs_diff)) / len(perm_replicates)

print('observed difference:',obs_diff)
print('p-value:',p_value)


# Calculating pearson correlations using stats.pearsonr
print(' ')
clean_df = clean_df[np.logical_not(clean_df['Excess Readmission Ratio'].isnull())]
print('pearson-r,   2-tailed-p-value')
print(stats.pearsonr(clean_df['Number of Discharges'], clean_df['Excess Readmission Ratio']))
```

I observe a difference between the two sets of 0.044.  P-value is nearly 0, meaning that in 10,000 permutations, none had a difference as great as our observed difference.  I will reject the Null Hypothesis, and conclude that the difference is statistically significant.
