### Background
Racial discrimination continues to be pervasive in cultures throughout the world. ### Data
In the dataset provided, each row represents a resume. The 'race' column has two values, 'b' and 'w', indicating black-sounding and white-sounding. The column 'call' has two values, 1 and 0, indicating whether the resume received a call from employers or not.

Note that the 'b' and 'w' values in race are assigned randomly to the resumes when presented to the employer.

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sns
import math
```

```python
data = pd.io.stata.read_stata('data/us_job_market_discrimination.dta')
```

```python
print('head: ', data.head())
print('tail: ', data.tail())
print('shape: ', data.shape)
print('info: ', data.info())
data.describe()
```

```python
# Extract from dataframe the relevant variables: race and callbacks
# Segregate into white and black
data_brief = data[['race', 'call']]
black = data_brief[data_brief.race == 'b']
white = data_brief[data_brief.race == 'w']

# Examine basic data, including percentage of callback rates for each group.
# number of callbacks for black-sounding names
print('# of callbacks for black-sounding names: ' , sum(black.call))
# number of callbacks for white-sounding names
print('# of callbacks for white-sounding names: ' , sum(white.call))

# number of black-sounding names
print('# of black-sounding names: ' , len(black))
# number of white-sounding names
print('# of white-sounding names: ' , len(white))

#fraction of callbacks for each group
black_rate = sum(black.call)/len(black)
white_rate = sum(white.call)/len(white)
print('callback rate for black-sounding names: ', black_rate)
print('callback rate for white-sounding names: ', white_rate)
print( 'Difference in callback rate: ', abs(black_rate - white_rate) )
print('White names were called back ', 
      100*((white_rate/black_rate)) ,'% as often as black names')
print('Black names were called back ',
     100*((black_rate/white_rate)), '% as often as white names')
```

```python
print(np.sum(white.call), len(white.call))
```

```python
# Null hypothesis that no significant difference between 2 arrays
# Calculate t statistic (percentage of difference of success with 2 arrays)
np.random.seed(697)

def t_stat(x,y):
    x_rate = np.sum(x) / len(x)
    y_rate = np.sum(y) / len(y)
    return(x_rate - y_rate)

obs_stat = t_stat(white.call, black.call)

merged_array = np.concatenate( (white.call, black.call), axis=0 )

perm_replicates = np.empty(10000)

# Create 10,000 permutation replicates
for i in range(10000):
    perm_resample = np.random.permutation(merged_array)
    white_perm = perm_resample[:len(white.call)]
    black_perm = perm_resample[len(white.call):]
    perm_replicates[i] = t_stat(white_perm,black_perm)

plt.hist(perm_replicates,bins=30,normed=True)
plt.xlabel('difference in callback rates')
plt.ylabel('count')
plt.tight_layout()
plt.show()

np.sort(perm_replicates)
print('95% Confidence Intervals:', np.percentile(perm_replicates,[2.5,97.5]))

# Margin of Error
standard = np.std(perm_replicates) / np.sqrt(len(perm_replicates))
print ('margin of error: ', standard*obs_stat)

# p-value
p_value = np.sum(perm_replicates >= obs_stat) / len(perm_replicates)
print('P-value: ', p_value)
```

**Examining Racial Discrimination in the U.S. job market**

1. a. What test is appropriate?  
I'll do a hypothosis test using permutation sampling.
1. b. Does CLT apply?  
Sample size is 2435 per group, sufficient to apply the Central Limit Theorem.

2. What are the null and alternate hypothoses?
Null H: no difference in callback rates in the general population between the 2 groups.
H1: significant difference in callback rates for black-sounding names, compared to white sounding names.

3. Compute a margin of error, confidence interval, and p-value.
I found 95% confidence interval between -0.015 and +0.014, with a very small margin of error, and a p-value very close to zero, meaning almost no chance of the sampled data in a population with no bias.  

4. Racial Discrimination.
